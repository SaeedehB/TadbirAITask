```python
import pandas as pd
import numpy as np
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
#nltk.download('stopwords')
#nltk.download('vader')
#nltk.download('punkt')
```

```python
list_of_entities = []
with open('./data/Entities.csv', 'r') as f:
    entities_info = f.readlines()
for entity_info in entities_info[1:]:
    entity_word = entity_info.split(',')[1]
    list_of_entities.append(entity_word)
```

```python
with open('./data/News.csv','rb') as f:
    news = f.readlines()
```

```python
sid = SentimentIntensityAnalyzer()
stopset = set(stopwords.words('english'))
for index,new in enumerate(news):
    sentence = new.decode('windows-1252')
    sentence = sentence.replace('-', ' ')
    sentence_tokenized =[word for word in word_tokenize(sentence)]
    entities_index = []
    entities_unique = []
    for ind,word in enumerate(sentence_tokenized):
        if(word in list_of_entities):
            entities_index.append(ind)
            entities_unique.append(word)
    
    entities_unique = set(entities_unique)
    sentiment_df = pd.DataFrame()
    if len(entities_unique) == 1:
        print(sentence_tokenized[entities_index[0]], sid.polarity_scores(sentence))
    if len(entities_unique) > 1:
        for i in range(len(entities_index)):
            try:
                start_index = entities_index[i]
                end_index = entities_index[i+1]
            except:
                start_index = entities_index[i]
                end_index = -1
            
            print(sentence_tokenized[entities_index[i]], sid.polarity_scores(' '.join(sentence_tokenized[start_index:end_index])))
```
